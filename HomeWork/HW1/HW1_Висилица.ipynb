```python
import random
import string
letters = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
MAX_ERRORS = 6

HANGMANPICS = ['''        
 _____
 |  ||
    ||
    ||
    ||
 ___||___ ''', '''

 _____
 |  || 
 o  ||
    ||
    ||
 ___||___ ''', '''
 _____
 |  ||
 o  ||
 |  ||
    ||
 ___||___ ''', '''
 _____
 |  ||
 o  ||
/|  ||
    ||
 ___||___ ''', '''
 _____
 |  ||
 o  ||
/|\ ||
    ||
 ___||___ ''', '''
 _____
 |  ||
 o  ||
/|\ ||
/   ||
 ___||___ ''', '''
 _____
 |  ||
 o  ||
/|\ ||
/ \ ||
 ___||___  ''']


def game_words():
    themes = ["цветы", "фрукты", "морфологические термины"]
    inp = True
    while inp: # работает до тех пор пока пользовательн не введет верную тему
        theme = input('Выберите одну из тем игры : цветы, фрукты, морфологические термины ').strip().lower()
        if theme not in themes:
            print('Такой темы нет, попробуйте еще раз')
        else:
            break
    game_file = theme + '.txt' #  название файла со словами

    with open(game_file, encoding = 'utf-8') as f: #  открывает файл со словами
        raw_words = f.read()
    words = raw_words.split()
    game_word = random.choice(words).strip().lower() #выбирает рандомное слово
    print('У вас есть 6 попыток, чтобы угадать слово длиной', len(game_word), 'букв', HANGMANPICS[0])
    return game_word

def letter_input():
    inp = True
    while inp: #  работает пока пользователь не введет букву
        letter = input('Введите одну букву: ').strip().lower()
        if len(letter) == 1:
            if letter not in letters:
                print('\nЭто должна быть буква!')
                break
            inp = False
        else:
            print('\nБуква должна состоять только из одного сивола!')
    return letter

def draw_field(letter, word, quessed):
    print('\n', ' ' * 10, end=' ')
    for i in word:
        if i in quessed:
            print(i, end=' ')
        elif i == letter:
            print(letter, end=' ')
        else:
            print('_', end=' ')
    print('\n')


def is_game_finished(errors_num, word, quessed):
    if errors_num == MAX_ERRORS:
        return True

    if set(word) == set(quessed):
        return True

    return False


def main():
    game = True
    while game:

        word = game_words()
        guessed_letters = []
        error_count = 0

        while not is_game_finished(error_count, word, guessed_letters): #работает пока игра не закончена
            letter = letter_input()
            if letter in word:
                guessed_letters.append(letter)
            else:
                error_count += 1
                if MAX_ERRORS-error_count == 1:
                    print('Буквы "{0}" нет. У вас осталась {1} попытка '.format(letter, MAX_ERRORS - error_count), HANGMANPICS[5])
                elif MAX_ERRORS-error_count == 2:
                    print('Буквы "{0}" нет. У вас осталось {1} попытки '.format(letter, MAX_ERRORS - error_count), HANGMANPICS[4])
                elif MAX_ERRORS-error_count == 3:
                    print('Буквы "{0}" нет. У вас осталось {1} попытки '.format(letter, MAX_ERRORS - error_count), HANGMANPICS[3])
                elif MAX_ERRORS-error_count == 4:
                    print('Буквы "{0}" нет. У вас осталось {1} попытки '.format(letter, MAX_ERRORS - error_count), HANGMANPICS[2])
                elif MAX_ERRORS-error_count == 5:
                    print('Буквы "{0}" нет. У вас осталось {1} попыток '.format(letter, MAX_ERRORS - error_count), HANGMANPICS[1])
                elif MAX_ERRORS-error_count == 6:
                    print('Буквы "{0}" нет. У вас осталось {1} попыток '.format(letter, MAX_ERRORS - error_count), HANGMANPICS[0])


            draw_field(letter, word, guessed_letters)

        if error_count < MAX_ERRORS:
            print('Поздравляем. Вы выиграли! с:')

        else:
            print('Простите, но вы проиграли, загаданное слово - "{0}"'.format(word), HANGMANPICS[6])
        again = input('Хотиете сыграть еще раз? Введите "да" или "нет" ')
        print(again)
        if again == 'да':
            continue
        elif again == 'нет':
            print('Было весело!')
            break
        else:
            print('Неверный ответ, будем считать, что "нет"')
            break


if __name__ == '__main__':
    main()




```
